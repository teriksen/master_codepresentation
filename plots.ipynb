```python
def read_data(file_name): # Read data from file and return as pandas dataframe
    if not os.path.exists(file_name):
        print(f"File '{file_name}' does not exist.")
        return None
    filetype = file_name.split('.')[-1]
    if filetype == 'csv':
        data = pd.read_csv(file_name)
    elif filetype == 'feather':
        data = pd.read_feather(file_name)
    else:
        print('Filetype not supported')
        return None
    return data

def get_3D_dims(data): # Get dimensions of 3D data (time, x, y)
    n_time_steps = len(data)
    n_x = np.round(np.sqrt(len(data.columns))).astype(int)
    n_y = n_x
    return n_time_steps, n_x, n_y

def convert_solution_data(solution): # Read solution data and return as 3D numpy array
    n_time_steps, n_x, n_y = get_3D_dims(solution)

    solution_3D = np.zeros((n_time_steps, n_x, n_y))

    for i in range(n_time_steps):
        solution_3D[i] = solution.iloc[i].values.reshape(n_x, n_y)

    return solution_3D

def get_reference_gzip_size(L):
    # read the file size of "data/angvel_series/" + L + "/feather/zip_size_reference.feather"
    bytes_size = os.path.getsize("data/angvel_series/" + L + "/feather/zip_size_reference.feather")
    return bytes_size

def convert_1D_data(data): # For shannon, file_sizes, energy
    assert len(data.columns) == 1, "Data should be 1D"
    n_time_steps = len(data)
    data_1D = np.zeros(n_time_steps)
    for i in range(n_time_steps):
        data_1D[i] = data.iloc[i].values[0]
    return data_1D

def convert_2D_data(data): # For orderparams, returns two 1D arrays
    assert len(data.columns) == 2, "Data should be 2D"
    n_time_steps = len(data)
    data_2D = np.zeros((n_time_steps, 2))
    for i in range(n_time_steps):
        data_2D[i] = data.iloc[i].values
    return data_2D[:,0], data_2D[:,1] # Return two 1D arrays

def animate_solution(solution_3D): # Animate 3D solution data
    # Create a figure and axis
    fig, ax = plt.subplots()
    plot_sol = solution_3D/(2*np.pi)
    n_time_steps, n_x, n_y = get_3D_dims(plot_sol)

    # assert that all values of plot_sol are between 0 and 1
    print("LIMITS OF THE DATA")
    print(np.min(plot_sol))
    print(np.max(plot_sol))

    # Set up the initial frame
    im = ax.imshow(plot_sol[0], cmap=cm.twilight, extent=[0, n_x, 0, n_y], vmin=0, vmax=1, animated=True)

    # Add a colorbar
    cbar = plt.colorbar(im, ax=ax)
    cbar.set_label("Orientation (radians)")

    # Update function for animation
    def update(frame):
        im.set_array(plot_sol[frame])
        return [im]

    # Create the animation
    ani = animation.FuncAnimation(fig, update, frames=range(0, n_time_steps, 2), interval=50, blit=True)

    # Use IPython's HTML display to show the animation
    HTML(ani.to_jshtml())

def cast_and_encode(solution, bits=8): # Cast solution to integers and encode as binary
    dim = 2**bits
    solution_binned = np.floor(dim * solution / (2*np.pi))
    solution_binned[solution_binned < 0] += dim
    
    # ensure that no entries exceed the bin size
    if np.any(solution_binned >= dim):
        print("Warning: Some entries are greater than the bin size.")
    if np.any(solution_binned < 0):
        print("Warning: Some entries are less than 0.")

    return solution_binned.astype(int)

def compare_mean_file_size_USELESS():
    system_sizes = [4,8,16,32,64]
    fig, ax = plt.subplots(figsize=(8, 6))


    for i, L in enumerate(system_sizes):
        angvels = angs[str(L)]
        color_shade = 0.05 + i/(len(system_sizes)+1)
        means = np.zeros(len(angvels))
        for j, ang in enumerate(angvels):
            file_sizes = read_data("data/angvel_series/" + str(L) + "/feather/" + ang + "w_file_sizes.feather")
            if file_sizes is None:
                continue
            file_sizes = convert_1D_data(file_sizes)
            means[j] = np.mean(file_sizes[-400:])
        peak = np.max(means)
        for j, ang in enumerate(angvels):
            if j == 0:
                ax.scatter(float(ang), means[j]/peak, label = str(L), color = cm.magma(color_shade), s = 5)
            else:
                ax.scatter(float(ang), means[j]/peak, color = cm.magma(color_shade), s = 5)
    ax.set_xlabel(r"$\Omega$", fontsize=14)
    ax.set_ylabel("Average file size (bytes)", fontsize=14)

    ax.legend()
    plt.show()

def get_phase_probability_distribution(system_size, L, time_steps, angvel, bits=8, std=1):
    n_bins = 2**bits
    prob_dist = np.zeros((time_steps, n_bins))
    orderparams = read_data("data/angvel_series/" + str(L) + "/feather/" + angvel + "w_orderparams.feather")
    if orderparams is None:
        print("Error: orderparams data not found.")
        return None
    orderparams = convert_2D_data(orderparams)
    avg_theta = orderparams[1]
    # check that all avg_theta has values between 0 and 2*pi
    if np.any(avg_theta < 0) or np.any(avg_theta > 2*np.pi):
        print("Warning: Some avg_theta values are out of bounds.")
        print("Minimum value:", np.min(avg_theta))
        print("Maximum value:", np.max(avg_theta))
        ValueError("avg_theta values out of bounds")
    avg_theta_binned = np.floor(n_bins * avg_theta / (2*np.pi))
    # make a discrete normal distribution around the mean
    for i in range(time_steps):
        # make a bounded normal distribution
        prob_dist_i = np.exp(-(np.arange(n_bins) - avg_theta_binned[i])**2 / (2*std**2))
        prob_dist_i /= np.sum(prob_dist_i)
        #if i == 900:
        #    print("prob_dist_i", prob_dist_i)
        prob_dist[i] = prob_dist_i
    return prob_dist
#p_l4_w10 = get_phase_probability_distribution("4", 4, 1000, "1.0", bits=8, std=2**6/1.5)
```

```python
def plot_shannon_coin_flip_example():
    fig, ax = plt.subplots(figsize=(8, 6))

    ps = np.linspace(0, 1, 101)
    heads_heads = ps * ps
    heads_tails = ps * (1 - ps)
    tails_heads = heads_tails
    tails_tails = (1 - ps) * (1 - ps)
    shannon_entropy = - (heads_heads * np.log2(heads_heads + 1e-10) + 
                         heads_tails * np.log2(heads_tails + 1e-10) + 
                         tails_heads * np.log2(tails_heads + 1e-10) + 
                         tails_tails * np.log2(tails_tails + 1e-10))
    ax.plot(ps, shannon_entropy, color=cm.magma(0.1), label=r"$S(p)$", linewidth=2)
    ax.plot(ps, -heads_heads*np.log2(heads_heads + 1e-10), color=cm.magma(0.3), label=r"$s(H,H)$", alpha=0.6)
    ax.plot(ps, -heads_tails*np.log2(heads_tails + 1e-10), color=cm.magma(0.5), label=r"$s(H,T)$, $s(T,H)$", alpha=0.6)
    ax.plot(ps, -tails_tails*np.log2(tails_tails + 1e-10), color=cm.magma(0.7), label=r"$s(T,T)$", alpha=0.6)

    print(ps[50], ps[90], ps[99])
    ax.set_xlabel(r"$p$")
    ax.set_xticks([0, 0.25, 0.5, 0.75, 1])
    ax.axvline(0.5, ymax=shannon_entropy[50]/np.max(shannon_entropy)-0.04, color='gray', linestyle='--', alpha=0.7)
    ax.axvline(0.9, ymax=shannon_entropy[90]/np.max(shannon_entropy), color='gray', linestyle='--', alpha=0.7)
    ax.axvline(1, ymax=shannon_entropy[99]/np.max(shannon_entropy)-0.03, color='gray', linestyle='--', alpha=0.7)
    ax.set_ylabel("S")
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.set_title("Shannon Entropy in a Coin Flip", fontsize=16)
    ax.legend(fontsize=12, loc='upper right')
    
    plt.savefig("figurer/shannon_coin_flip_example.png", bbox_inches='tight', dpi=500)
    plt.show()

#plot_shannon_coin_flip_example()
```

```python
def remove_outliers(data, outliers, system_sizes, label=""):
    print(f"\n--- Removing {label} outliers ---")
    for sys_size, series_index in outliers:
        try:
            sys_index = system_sizes.index(sys_size)
            if 0 <= series_index - 1 < len(data[sys_index]):
                removed = data[sys_index].pop(series_index - 1)
                print(f"Removed {removed} from system size {sys_size}, series {series_index}")
            else:
                print(f"Index {series_index} out of range for system size {sys_size}")
        except ValueError:
            print(f"System size {sys_size} not found in system_sizes")
        except Exception as e:
            print(f"Unexpected error: {e}")

def plot_avg_crit_points(crit_points_r, crit_points_s, crit_points_gzip, system_sizes= [8, 12, 16, 20, 24, 32, 40, 48, 56], save_fig=False,
                         r_outliers=[], shannon_outliers=[], gzip_outliers=[]):
    fig, axs = plt.subplots(1,3,figsize=(11, 6), sharey=True)
    
    crit_points_r = crit_points_r.tolist()
    crit_points_s = crit_points_s.tolist()
    crit_points_gzip = crit_points_gzip.tolist()
    remove_outliers(crit_points_r, r_outliers, system_sizes, label="r")
    remove_outliers(crit_points_s, shannon_outliers, system_sizes, label="shannon")
    remove_outliers(crit_points_gzip, gzip_outliers, system_sizes, label="gzip")
    avg_crit_points_r = np.array([np.mean(cp) for cp in crit_points_r])
    std_crit_points_r = np.array([np.std(cp) for cp in crit_points_r])
    avg_crit_points_s = np.array([np.mean(cp) for cp in crit_points_s])
    std_crit_points_s = np.array([np.std(cp) for cp in crit_points_s])
    avg_crit_points_gzip = np.array([np.mean(cp) for cp in crit_points_gzip])
    std_crit_points_gzip = np.array([np.std(cp) for cp in crit_points_gzip])
    
    print("Average critical points (r):", avg_crit_points_r)
    print("Average critical points (shannon):", avg_crit_points_s)
    print("Average critical points (gzip):", avg_crit_points_gzip)
    axs[0].errorbar(system_sizes, avg_crit_points_r, yerr=std_crit_points_r, fmt='o-', color=cm.magma(0.3), capsize=4)
    axs[1].errorbar(system_sizes, avg_crit_points_s, yerr=std_crit_points_s, fmt='o-', color=cm.magma(0.5), capsize=4)
    axs[2].errorbar(system_sizes, avg_crit_points_gzip, yerr=std_crit_points_gzip, fmt='o-', color=cm.magma(0.7), capsize=4)
    axs[0].set_title(r'$\widebar\Omega_c$ from $\Delta r$', fontsize=16)
    axs[1].set_title(r'$\widebar\Omega_c$ from $\Delta S$', fontsize=16)
    axs[2].set_title(r'$\widebar\Omega_c$ from $\Delta G$', fontsize=16)
    axs[0].set_ylabel(r'$\Omega_c(L)$', fontsize=15)
    fig.suptitle(r'Average critical values of $\Omega$ for different system sizes', fontsize=20)
    for ax in axs:
        #ax.set_xticks(np.arange(8, 57, 8))
        ax.set_xticklabels([str(size) for size in system_sizes], fontsize=14)
        ax.set_xlabel('System Size (L)', fontsize=15)
        ax.grid(True, linestyle='--', alpha=0.5)
    axs[0].set_yticks(np.arange(0, 3.1, 0.5))
    axs[0].set_yticklabels([f"{i:.1f}" for i in np.arange(0, 3.1, 0.5)], fontsize=14)
    if save_fig:
        plt.savefig("figurer/avg_critical_frequencies.png", dpi=500, bbox_inches='tight')
        print("Saved figure to figurer/avg_critical_frequencies.png")
    plt.tight_layout()
    plt.show()
    return avg_crit_points_r, avg_crit_points_s, avg_crit_points_gzip, std_crit_points_r, std_crit_points_s, std_crit_points_gzip
crit_ps_r = crit_points.copy()
crit_ps_shannon = crit_points_shannon.copy()
crit_ps_gzip = crit_points_gzip.copy()
acp_r, acp_s, acp_gz, std_r, std_s, std_gz = plot_avg_crit_points(crit_ps_r, crit_ps_shannon, crit_ps_gzip, save_fig=True, 
                                                                  #gzip_outliers=[(48,2), (48,5)], shannon_outliers=[(16, 2)], r_outliers=[(16, 2), (20,5)]
                                                                  )
```

```python
def plot_phase_of_single_rotor(savefig=False):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax2 = ax.twinx()
    angvel = "0.3"
    sys = "64anim_rand"
    path = "data/angvel_series/" + sys + "/feather/" + angvel + "w_rotorhist.feather"
    if not os.path.exists(path):
        print("File does not exist: ", path)
        return None
    rotorhist = read_data(path)
    rotorhist = convert_solution_data(rotorhist)
    ax.plot(rotorhist[:, 0, 0], color=cm.magma(0.3), label=r'$\theta(t)$')
    ax2.plot(np.cos(rotorhist[:, 0, 0]), color=cm.magma(0.7), label=r'$\cos(\theta(t))$')
    # calculate the period of the oscillation
    t_start = 500
    t_end = 4000
    # find first top
    max, max_index = 0, 0
    for i in range(t_start, 2000):
        if np.cos(rotorhist[i, 0, 0]) > max:
            max = np.cos(rotorhist[i, 0, 0])
            max_index = i
    # find second top
    print(f"First top found at index {max_index} with value {max:.2f}")
    max2, max_index2 = 0, 0
    for i in range(max_index+1, t_end):
        if np.cos(rotorhist[i, 0, 0]) > max2:
            max2 = np.cos(rotorhist[i, 0, 0])
            max_index2 = i
    print(f"Second top found at index {max_index2} with value {max2:.2f}")
    period = max_index2 - max_index
    print(f"Period of oscillation: {period:.2f} time steps")
    
    plt.annotate('', xy=(max_index, 1), xytext=(max_index2, 1),
             arrowprops=dict(arrowstyle='<->', color='black'))

    plt.text((max_index + max_index2) / 2, 1 - 0.1, rf'{int(period)} time units',
             ha='center', va='top', color='black', fontsize=14)

    ax.set_xlabel(r'Time $(t)$', fontsize=16)
    ax.set_ylabel(r'Phase $(\theta)$', fontsize=16)
    ax2.set_ylabel(r'$\cos(\theta)$', fontsize=16)
    ax.set_yticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_yticklabels([r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'], fontsize=14)
    ax2.set_yticks([-1, -0.5, 0, 0.5, 1])
    ax2.set_yticklabels([r'$-1$', r'$-0.5$', r'$0$', r'$0.5$', r'$1$'], fontsize=14)
    ax.set_title(r'Phase history of a single rotor.  $L=64$, $\Omega=0.3$', fontsize=18)
    legends = ax.get_legend_handles_labels()
    legends2 = ax2.get_legend_handles_labels()
    ax.legend(legends[0] + legends2[0], legends[1] + legends2[1], 
              #loc=(0.5, 0.85), 
              fontsize=14)
    ax.grid(True, linestyle='--', alpha=0.5)
    if savefig:
        plt.savefig("figurer/single_rotor_phase_L64_w" + angvel + ".png", dpi=500, bbox_inches='tight')
        print("Figure saved as 'single_rotor_phase_L64_w" + angvel + ".png'")
    plt.show()

#plot_phase_of_single_rotor(savefig=True)
```

```python
def compare_phase_velocities_64_0p7(savefig=False):
    fig, ax = plt.subplots(figsize=(7, 5), sharex=True)
    path_presynch = "data/angvel_series/64anim_presynch/feather/0.7w_rotorhist.feather"
    path_rand = "data/angvel_series/64anim_rand/feather/0.7w_rotorhist.feather"
    rotorhist_presynch = read_data(path_presynch)
    rotorhist_rand = read_data(path_rand)
    rotorhist_presynch = convert_solution_data(rotorhist_presynch)
    rotorhist_rand = convert_solution_data(rotorhist_rand)
    if not os.path.exists(path_presynch):
        print("File does not exist: ", path_presynch)
        return None
    if not os.path.exists(path_rand):
        print("File does not exist: ", path_rand)
        return None
    Dt = 2
    start_pad = 100
    x_ticks = np.arange(0, 4001, 500)
    x_ax = np.arange(start_pad+Dt, rotorhist_presynch.shape[0])
    ax.set_xticks(x_ticks)
    ax.set_xticklabels([str(i) for i in range(0, 4001, 500)], fontsize=16)
    y_ticks = np.linspace(0, 0.015, 6)
    ax.set_ylim(-0.0005, 0.0155)
    ax.set_yticks(y_ticks)
    ax.set_yticklabels([f"{tick:.3f}" for tick in y_ticks], fontsize=16)

    rotorhist_diff_presynch = np.angle(np.exp(1j * (rotorhist_presynch[start_pad+Dt:, :, :] - rotorhist_presynch[start_pad:-Dt, :, :])))
    rotorhist_diff_rand = np.angle(np.exp(1j * (rotorhist_rand[start_pad+Dt:, :, :] - rotorhist_rand[start_pad:-Dt, :, :])))
    avg_phase_change_presynch = np.mean(np.abs(rotorhist_diff_presynch), axis=(1, 2))/Dt
    avg_phase_change_rand = np.mean(np.abs(rotorhist_diff_rand), axis=(1, 2))/Dt
    ax.plot(x_ax, avg_phase_change_rand, color=cm.magma(0.3), label=r'$\langle \Delta \theta \rangle$', linewidth=2)
    ax.plot(x_ax, avg_phase_change_presynch, color=cm.magma(0.7), label=r'$\langle \Delta \theta \rangle$, pre-synchronized', linewidth=2)
    ax.set_xlabel(r'Time $(t)$', fontsize=18)
    ax.set_ylabel(r'$\langle \Delta \theta \rangle$', fontsize=18)
    ax.set_title(r'$\langle \Delta \theta \rangle$ over a unit time step $\text{D}t=1$: $\Omega=0.7$', fontsize=22)
    ax.legend(fontsize=16)
    ax.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    if savefig:
        plt.savefig("figurer/compare_phase_velocities_64_0p7.png", dpi=500, bbox_inches='tight')
        print("Figure saved as 'compare_phase_velocities_64_0p7.png'")
    plt.show()

#compare_phase_velocities_64_0p7(savefig=True)
```
