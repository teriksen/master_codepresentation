```python
def read_data(file_name): # Read data from file and return as pandas dataframe
    if not os.path.exists(file_name):
        print(f"File '{file_name}' does not exist.")
        return None
    filetype = file_name.split('.')[-1]
    if filetype == 'csv':
        data = pd.read_csv(file_name)
    elif filetype == 'feather':
        data = pd.read_feather(file_name)
    else:
        print('Filetype not supported')
        return None
    return data

def get_3D_dims(data): # Get dimensions of 3D data (time, x, y)
    n_time_steps = len(data)
    n_x = np.round(np.sqrt(len(data.columns))).astype(int)
    n_y = n_x
    return n_time_steps, n_x, n_y

def convert_solution_data(solution): # Read solution data and return as 3D numpy array
    n_time_steps, n_x, n_y = get_3D_dims(solution)

    encoded_solution_3D = np.zeros((n_time_steps, n_x, n_y))

    for i in range(n_time_steps):
        encoded_solution_3D[i] = solution.iloc[i].values.reshape(n_x, n_y)

    return encoded_solution_3D

def get_reference_gzip_size(L):
    # read the file size of "data/angvel_series/" + L + "/feather/zip_size_reference.feather"
    bytes_size = os.path.getsize("data/angvel_series/" + L + "/feather/zip_size_reference.feather")
    return bytes_size

def convert_1D_data(data): # For shannon, file_sizes, energy
    assert len(data.columns) == 1, "Data should be 1D"
    n_time_steps = len(data)
    data_1D = np.zeros(n_time_steps)
    for i in range(n_time_steps):
        data_1D[i] = data.iloc[i].values[0]
    return data_1D

def convert_2D_data(data): # For orderparams, returns two 1D arrays
    assert len(data.columns) == 2, "Data should be 2D"
    n_time_steps = len(data)
    data_2D = np.zeros((n_time_steps, 2))
    for i in range(n_time_steps):
        data_2D[i] = data.iloc[i].values
    return data_2D[:,0], data_2D[:,1] # Return two 1D arrays

def animate_solution(solution_3D): # Animate 3D solution data
    # Create a figure and axis
    fig, ax = plt.subplots()
    plot_sol = solution_3D/(2*np.pi)
    n_time_steps, n_x, n_y = get_3D_dims(plot_sol)

    # assert that all values of plot_sol are between 0 and 1
    print("LIMITS OF THE DATA")
    print(np.min(plot_sol))
    print(np.max(plot_sol))

    # Set up the initial frame
    im = ax.imshow(plot_sol[0], cmap=cm.twilight, extent=[0, n_x, 0, n_y], vmin=0, vmax=1, animated=True)

    # Add a colorbar
    cbar = plt.colorbar(im, ax=ax)
    cbar.set_label("Orientation (radians)")

    # Update function for animation
    def update(frame):
        im.set_array(plot_sol[frame])
        return [im]

    # Create the animation
    ani = animation.FuncAnimation(fig, update, frames=range(0, n_time_steps, 2), interval=50, blit=True)

    # Use IPython's HTML display to show the animation
    HTML(ani.to_jshtml())

def cast_and_encode(solution, bits=8): # Cast solution to integers and encode as binary
    dim = 2**bits
    solution_binned = np.floor(dim * solution / (2*np.pi))
    solution_binned[solution_binned < 0] += dim
    
    # ensure that no entries exceed the bin size
    if np.any(solution_binned >= dim):
        print("Warning: Some entries are greater than the bin size.")
    if np.any(solution_binned < 0):
        print("Warning: Some entries are less than 0.")

    return solution_binned.astype(int)

def compare_mean_file_size_USELESS():
    system_sizes = [4,8,16,32,64]
    fig, ax = plt.subplots(figsize=(8, 6))


    for i, L in enumerate(system_sizes):
        angvels = angs[str(L)]
        color_shade = 0.05 + i/(len(system_sizes)+1)
        means = np.zeros(len(angvels))
        for j, ang in enumerate(angvels):
            file_sizes = read_data("data/angvel_series/" + str(L) + "/feather/" + ang + "w_file_sizes.feather")
            if file_sizes is None:
                continue
            file_sizes = convert_1D_data(file_sizes)
            means[j] = np.mean(file_sizes[-400:])
        peak = np.max(means)
        for j, ang in enumerate(angvels):
            if j == 0:
                ax.scatter(float(ang), means[j]/peak, label = str(L), color = cm.magma(color_shade), s = 5)
            else:
                ax.scatter(float(ang), means[j]/peak, color = cm.magma(color_shade), s = 5)
    ax.set_xlabel(r"$\Omega$", fontsize=14)
    ax.set_ylabel("Average file size (bytes)", fontsize=14)

    ax.legend()
    plt.show()

def get_phase_probability_distribution(system_size, L, time_steps, angvel, bits=8, std=1):
    n_bins = 2**bits
    prob_dist = np.zeros((time_steps, n_bins))
    orderparams = read_data("data/angvel_series/" + str(L) + "/feather/" + angvel + "w_orderparams.feather")
    if orderparams is None:
        print("Error: orderparams data not found.")
        return None
    orderparams = convert_2D_data(orderparams)
    avg_theta = orderparams[1]
    # check that all avg_theta has values between 0 and 2*pi
    if np.any(avg_theta < 0) or np.any(avg_theta > 2*np.pi):
        print("Warning: Some avg_theta values are out of bounds.")
        print("Minimum value:", np.min(avg_theta))
        print("Maximum value:", np.max(avg_theta))
        ValueError("avg_theta values out of bounds")
    avg_theta_binned = np.floor(n_bins * avg_theta / (2*np.pi))
    # make a discrete normal distribution around the mean
    for i in range(time_steps):
        # make a bounded normal distribution
        prob_dist_i = np.exp(-(np.arange(n_bins) - avg_theta_binned[i])**2 / (2*std**2))
        prob_dist_i /= np.sum(prob_dist_i)
        #if i == 900:
        #    print("prob_dist_i", prob_dist_i)
        prob_dist[i] = prob_dist_i
    return prob_dist
#p_l4_w10 = get_phase_probability_distribution("4", 4, 1000, "1.0", bits=8, std=2**6/1.5)
```

```python
def calculate_period(full_rs, epsilon=0.005):
    rs = full_rs[400:600] # We assume the system has reached a steady state after t=600
    max_r = np.max(rs)
    argmax_r = np.argmax(rs)
    # Find index of nearest local minimum to the right of the maximum
    next_local_minimum = rs[argmax_r + 1]
    next_local_minimum_index = argmax_r + 1
    for i in range(argmax_r + 10, len(rs)):
        if rs[i] > max_r - epsilon:
            break
        if rs[i-1] > rs[i] < rs[i+1] and rs[i] < next_local_minimum:
            next_local_minimum = rs[i]
            next_local_minimum_index = i
    local_min_index = next_local_minimum_index - argmax_r
    next_argmax_r = np.argmax(rs[argmax_r+local_min_index:]) + argmax_r + local_min_index
    next_next_argmax_r = np.argmax(rs[next_argmax_r+1:next_argmax_r+2*local_min_index]) + next_argmax_r + 1
    print("First argmax_r:", argmax_r)
    print("Next argmax_r:", next_argmax_r)

#calculate_period(test, epsilon=0.005)
```

```python
def L32_plot_phase_transition_with_critical_point(critical_point=0, save_fig=False, 
                                              l_limit = 0, r_limit = 5, filename = "", title = ""):
    system_size="32"
    r_skip = 0
    l_skip = 0
    for angvel in angs[system_size]:
        if float(angvel) > r_limit:
            r_skip += 1
        if float(angvel) < l_limit:
            l_skip += 1

    n_angvels = len(angs[system_size])
    angvels = angs[system_size][l_skip:n_angvels-r_skip]

    fig, ax = plt.subplots(figsize=(8, 6))
    r_std = np.zeros(len(angvels))
    for i, angvel in enumerate(angvels):
        path = "data/angvel_series/" + system_size + "/feather/" + angvel + "w_"
        orderparams = read_data(path + "orderparams.feather") #2D
        r_hist, theta_hist = convert_2D_data(orderparams)
        r_std[i] = np.std(r_hist[-600:])

    x_axis = [float(angvel) for angvel in angvels]
    ang_min = np.min(x_axis)
    ang_max = np.max(x_axis)
    if critical_point == 0:
        critical_point = x_axis[np.argmax(r_std)]
        ax.axvline(x=critical_point, color='red', linestyle='--', label=r'Critical point, $\Omega_c = ${}'.format(critical_point), alpha=0.4)
    elif critical_point == -1:
        pass
    else:
        ax.axvline(x=critical_point, color='red', linestyle='--', label=r'Critical point, $\Omega_c = ${}'.format(critical_point), alpha=0.4)
    if ang_max - ang_min >= 4:
        x_ticks = np.arange(np.floor(ang_min), np.ceil(ang_max)+1, 1).tolist() + [critical_point]
    elif ang_max - ang_min >= 3:
        min_tick = 0
        max_tick = 0
        while min_tick <= ang_min - 0.5:
            min_tick += 0.5
        while max_tick <= ang_max + 0.5:
            max_tick += 0.5
        x_ticks = np.arange(min_tick, max_tick, 0.5).tolist() + [critical_point]
    else:
        min_tick = 0
        max_tick = 0
        while min_tick <= ang_min - 0.25:
            min_tick += 0.25
        while max_tick <= ang_max + 0.25:
            max_tick += 0.25
        x_ticks = np.arange(min_tick, max_tick, 0.25).tolist() + [critical_point]
    ax.plot(x_axis, r_std, marker='o', color=cm.magma(0.35), label=r'$\Delta$r')
    
    if title == "":
        ax.set_title(r'Phase transition in 32x32 system: $\Delta$r versus $\Omega$', fontsize=20)
    else:
        ax.set_title(title, fontsize=20)
    ax.set_xlabel(r'$\Omega$', fontsize=16)
    ax.set_ylabel(r'$\Delta$r', fontsize=16)
    ax.grid(True, linestyle='--', alpha=0.5)
    #ax.axvline(x=1.3, color='red', linestyle='--', label='Something happens', alpha=0.5)
    # remove last element from x_ticks
    if critical_point in x_ticks:
        x_ticks.remove(critical_point)
    ax.set_xticks(x_ticks)
    x_tick_labels = [str(x) for x in x_ticks]
    #x_tick_labels[-1] = r"$\Omega_c$"
    ax.set_xticklabels(x_tick_labels, rotation=45, fontsize=16)
    y_tick_labels = [str(round(y, 2)) for y in ax.get_yticks()]
    ax.set_yticklabels(y_tick_labels, fontsize=16)
    ax.legend(loc='lower right', fontsize=16)

    print("Number of points: ", len(angvels))
    print("STD(r) peak: ", np.max(r_std), ", at w =", x_axis[np.argmax(r_std)])


    plt.tight_layout()
    if save_fig:
        if filename == "":
            plt.savefig("figurer/"+system_size+"x"+system_size+"_phase_transition.png", dpi=500, bbox_inches='tight')
            print("Saved figure to figurer/"+system_size+"x"+system_size+"_phase_transition.png")
        else:
            plt.savefig("figurer/"+filename, dpi=500, bbox_inches='tight')
            print("Saved figure to figurer/"+filename)
    plt.show()
    return critical_point

L32_plot_phase_transition_with_critical_point(save_fig=True, l_limit=0.50, r_limit=3, critical_point=1.42)
```

```python
from matplotlib.colors import LinearSegmentedColormap
def example_domain_walls():

    # Create grid
    n_points = 25
    x, y = np.meshgrid(np.linspace(-5, 5, n_points), np.linspace(-5, 5, n_points))

    # Define vertical vector field with transition across x=0
    v = -np.tanh(x)       # From +1 to -1 across domain wall
    u = np.zeros_like(v)  # No horizontal component

    # Normalize v to [0, 1] for color mapping
    colors = (v + 1) / 2

    # Define custom colormap: red -> black -> blue
    custom_cmap = LinearSegmentedColormap.from_list("red_black_blue", ["blue", "black", "red"])

    
    # Plot
    plt.figure(figsize=(6, 4))
    # zoom in
    plt.quiver(x, y, u, v, colors, cmap=custom_cmap, pivot='middle', scale=20,
               headlength=4, headaxislength=3, width=0.01, alpha=0.8)
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.title("Example of a Domain Wall", fontsize=18)
    plt.xlabel("x", fontsize=16)
    plt.ylabel("y", fontsize=16)
    #plt.axis('equal')
    plt.grid(True)
    #plt.colorbar(label='Orientation (up → down)', cmap=custom_cmap)
    # set colorbar ticks
    plt.clim(0, 1)
    # increase font size of colorbar
    cbar = plt.colorbar()
    cbar.ax.tick_params(labelsize=12)
    cbar.set_label('Normalized Orientation', fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.savefig("figurer/example_domain_wall.png", dpi=500, bbox_inches='tight')
    plt.show()

#example_domain_walls()
```
